```python
df = pd.DataFrame({'service':df1.service, 'org':df1.org, 'explain':df1.explain})
df.head()
```

```python
idx = []
i = 0

ee = []
oo = []
ss = []
for e, o, s in zip(df['explain'], df['org'], df['service']):
    e, o, s = ast.literal_eval(e), ast.literal_eval(o), ast.literal_eval(s)
    
    if (len(e) != len(o)) or (len(e) != len(s)) or (len(o) != len(s)):
        idx.append(i)
        i += 1
    
    ee.append(e)
    oo.append(o)
    ss.append(s)
```
